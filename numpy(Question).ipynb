### Part 1: Basic Array Operations

```python
import numpy as np
arr = np.array([5, 10, 15, 20, 25, 30])
print(arr.sum())
print(arr.max())
print(arr.min())
print(arr.argmax())
print(arr.argmin())
print(arr.mean())
print(np.median(arr))
print(np.var(arr))
print(np.std(arr))
```
